```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
from torch.utils.data import DataLoader
from tqdm import tqdm
import matplotlib.pyplot as plt
import datetime

import custom_dataset as ds

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
def train_model(model, dataset, epoch, batch_size=16, lr=0.01, decay=0.95):
    criterion = torch.nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
    scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer, 0.95)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

    model.train()
    for epoch in range(epoch):        
        epoch_loss = 0
        for X, X_t, y, y_t in tqdm(dataloader):
            optimizer.zero_grad()
            X = X.to(model.device)
            X_t = X_t.to(model.device)
            
            y_pred = model((X, X_t))
            '''
            for i in range(dataset.next_step-1):
                y_pred = model(y_pred.unsqueeze(1).unsqueeze(1))'''
            loss = criterion(y_pred.cpu().squeeze(), y.squeeze())
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item()      

        scheduler.step()
        print(f'Epoch {epoch+1}, MSE Loss: {epoch_loss/len(dataloader)}')
```

```python
def inference_model(model, dataset, batch_size=16):
    with torch.no_grad():
        model.eval()
        dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

        ret = []
        for X, X_t in tqdm(dataloader):
            X = X.to(model.device)                  
            X_t = X_t.to(model.device)
            
            for i in range(dataset.next_step-1):
                y_pred = model((X, X_t))
                X = torch.cat([X[:, 1:], y_pred[:, :1]], dim=1)
                X_t = torch.remainder(torch.add(X_t, 1), TIME_LEN)
            y_pred = model((X, X_t))
            
            '''for i in range(dataset.next_step-1):
                y_pred = model(y_pred.unsqueeze(1).unsqueeze(1))'''
            ret.append(y_pred.cpu().squeeze())
            #ret.append(y_pred.cpu().squeeze())
    return ret
```

```python
TIME_LEN = 7*24*12

node_path = './data/processed_data/node_link/daegu_selected_nodes.csv'
link_path = './data/processed_data/node_link/daegu_selected_links.csv'
```

```python
class trash_layer(nn.Module):
    def __init__(self, device, mask):
        super(trash_layer, self).__init__()
        self.device = device
        self.mask = mask
        self.linear = nn.Linear(mask.shape[0], mask.shape[0], device=device, bias=True)
        
    def forward(self, x):
        with torch.no_grad():
            self.linear.weight.data = self.linear.weight.data * self.mask
        return self.linear(x)
        

class trash_net(nn.Module):
    def __init__(self, device, link_file, prev_len, hidden_channels, width):
        super(trash_net, self).__init__()
        self.device = device
        self.width = width             
        
        links = pd.read_csv(link_file)
        self.mask = pd.DataFrame(0, index = links['LINK_ID'], columns=links['LINK_ID'])

        for idx,row in links.iterrows():
            for idx2,row2 in links.iterrows():
                if (row['F_NODE'] == row2['F_NODE'] or row['F_NODE'] == row2['T_NODE'] or row['T_NODE'] == row2['F_NODE'] or row['T_NODE'] == row2['T_NODE']):
                    self.mask.loc[row['LINK_ID'], row2['LINK_ID']] = 1
        self.mask = torch.Tensor(self.mask.values).to(device)
        
        self.hidden_channels = hidden_channels
                 
        self.init_conv = nn.ModuleList([nn.Conv2d(prev_len*2, hidden_channels, 1, padding=0, device=device) for i in range(width)])
        #self.init_conv = nn.ModuleList([nn.Conv2d(prev_len, hidden_channels, 1, padding=0, device=device) for i in range(width)])
        self.init_trash = nn.ModuleList([trash_layer(device, self.mask) for i in range(width-1)])

        self.path_list = nn.ModuleList([nn.Sequential(*[nn.Sequential(self.init_trash[j], nn.ReLU()) for j in range(i)], self.init_conv[i], nn.ReLU(), \
                                        *[nn.Sequential(nn.Conv2d(hidden_channels, hidden_channels, 1, padding=0, device=device), nn.ReLU()) for j in range(self.width - i - 1)]) for i in range(self.width)])
        
        self.final_list = nn.Sequential(*[nn.Sequential(nn.Conv2d(hidden_channels * i, hidden_channels*(i-1), 1, padding=0, device=device), nn.ReLU()) for i in range(self.width, 1, -1)])
        
        self.conv_last = nn.Conv2d(hidden_channels, 1, 1, padding=0, device=device)
        
        self.time_embedd = nn.Embedding(TIME_LEN, len(links), device=device)

    def forward(self, x):
        x, xt = x
        #x += self.time_embedd(xt)
        x = torch.cat([x, self.time_embedd(xt)], dim=1)
        temp = [path(x) for path in self.path_list]
        temp = self.final_list(torch.cat(temp, dim=1))
        out = self.conv_last(temp)
        
        return out
```

```python
prev_len = 6
hidden_channels = 12
width = 4

md = trash_net(device, link_path, prev_len, hidden_channels, width)
print(md)
```

```python
train_data_path = './data/processed_data/traffic/2023_6to9.csv'
train_time_path = './data/processed_data/traffic/2023_6to9_t.csv'

train_next_step = 1

train_data = ds.Link_Time_Dataset(train_data_path, train_time_path, prev_len, train_next_step)
train_model(md, train_data, 25)
```

```python
def show_plot(d1, d2):
    fig, ax = plt.subplots(figsize=(20, 10))
    ax.plot(d1, 'r')
    ax.plot(d2, 'b')
    ax.grid(True)
    ax.set_xlabel('Time', loc='right')
    ax.set_ylabel('Speed', loc='top')
    plt.show()
```
